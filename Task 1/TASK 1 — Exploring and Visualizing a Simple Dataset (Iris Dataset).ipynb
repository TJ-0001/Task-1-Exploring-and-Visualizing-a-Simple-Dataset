### Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the Iris dataset

```python
df = pd.read_csv('Iris.csv')
```

### Print the shape, column names, and the first few rows using .head().

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

### Show basic info about the dataset

```python
df.info()
```

### Describe basic statistics of the dataset

```python
df.describe()
```

### Scatterplot to observe relationships between Sepal Length and Sepal Width by Species

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data = df,
    x = "SepalLengthCm",
    y = "SepalWidthCm",
    hue="Species",
    palette="dark",
)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

### Scatterplot to observe relationships between Petal Length and Petal Width by Species

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue="Species",
    palette="dark"
)
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```

### Histograms to show value distributions.

```python
data=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
data.hist(figsize=(10,8), bins = 15)
plt.suptitle("Feature Distributions (Histograms)")
plt.show()
```

## Boxplot to check for outliers and feature distribution

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
plt.title("Box Plot of Iris Features")
plt.xticks(rotation = 45)
plt.show()
```
